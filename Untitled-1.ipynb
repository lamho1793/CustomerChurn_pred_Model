```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
%matplotlib inline
```

```python
f=pd.read_csv('Data_jan.csv',low_memory=False)
```

```python
f[(f['是否离网']=='Y')]['本地活跃度'].fillna(0).sort_values().tolist()
```

```python
df=f[(f['是否离网']=='Y')][['过去3月当月本地活跃天数']].fillna(0)
df.sort_values(by=['过去3月当月本地活跃天数'],inplace=True)
df_mean = np.mean(df["过去3月当月本地活跃天数"])
df_std = np.std(df["过去3月当月本地活跃天数"])
pdf_Y = scipy.stats.norm.pdf(df["过去3月当月本地活跃天数"], df_mean, df_std)

df2=f[(f['是否离网']=='N')][['过去3月当月本地活跃天数']].fillna(0)
df2.sort_values(by=['过去3月当月本地活跃天数'],inplace=True)
df_mean2 = np.mean(df2["过去3月当月本地活跃天数"])
df_std2 = np.std(df2["过去3月当月本地活跃天数"])
pdf_N = scipy.stats.norm.pdf(df["过去3月当月本地活跃天数"], df_mean2, df_std2)

plt.plot(df["过去3月当月本地活跃天数"], pdf_Y, color='r')
plt.plot(df["过去3月当月本地活跃天数"], pdf_N, color='g')

plt.legend()
  
# To load the display window
plt.show()

# x=np.sort(f[(f['是否离网']=='Y')]['本地活跃度'].fillna(0).sort_values().tolist())
```
