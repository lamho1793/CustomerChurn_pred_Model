```python
col_dec=[]
for i in range(95):
    if (set(f[f.columns[i]].unique().tolist())=={'Y','N','0'} or set(f[f.columns[i]].unique().tolist())=={'Y','N'}) and f.columns[i]!='是否离网':
        col_dec.append(f.columns[i])
        print(f.columns[i])
```

```python
for x in col_dec:
        print(x)
        a=pd.get_dummies(f[x]).astype(int)
        if a.shape[1]==3:
            if x in ['auto_ext_ind', 'wth_auto_ext_ofr','是否携号入网','是否第一张合约届满','是否自动续约']:
                b=a['N'].replace(1,int(-1))+a['0'].replace(1,int(0))+a['Y']
                f.at[:,[x]]=b.to_frame(x)
            else:
                b=a['Y'].replace(1,int(-1))+a['0'].replace(1,int(0))+a['N']
                f.at[:,[x]]=b.to_frame(x)
        elif a.shape[1]==2:
            if 'auto_ext_ind' in ['auto_ext_ind', 'wth_auto_ext_ofr','是否携号入网','是否第一张合约届满','是否自动续约']:
                b=a['N'].replace(1,int(-1))+a['Y']
                f.at[:,[x]]=b.to_frame(x)
            else:
                b=a['Y'].replace(1,int(-1))+a['N']
                f.at[:,[x]]=b.to_frame(x)
```
